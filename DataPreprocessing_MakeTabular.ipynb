# NXT DATASET

```python
f = open("nxt.xml","r")
```

```python
lines = f.readlines()
```

```python
# lines
```

```python
len(lines)
```

```python
import numpy as np
```

```python
arr = []
```

```python
for line in lines:
  if line.startswith("<break"):
    index = lines.index(line)
    nextline = lines[index+1]
    linesubs = line.split(' ')
    nextlinesubs = nextline.split(' ')
    id = linesubs[1].split("=")[1].split("\"")[1].split("\"")[0]
    href = nextlinesubs[1].split("=")[1].split("\"")[1].split("\"")[0]
    start = linesubs[2].split("=")[1].split("\"")[1].split("\"")[0]
    end = linesubs[3].split("=")[1].split("\"")[1].split("\"")[0]
    index = (linesubs[5].split("=")[1].split(">")[0]).split("\"")[1].split("\"")[0]
    entry = [id, href, start,end,index]
    # print(entry)
    arr.append(entry)




arr[1]
```

```python
import pandas as pd

df = pd.DataFrame(arr, columns = ['id','href','start','end','index'])

df
```

```python
# xlr = pd.ExcelWriter('/content/drive/MyDrive/DataPreprocessingResults.xlsx')
df.to_excel( 'output.xlsx')
```

# BURSC

```python
f = open("bursc.brk","r")
lines = f.readlines()
lines
```

```python
arr = []
```

```python
for idx, line in enumerate(lines):
  if line.startswith(" "):
    line.strip()
    linesubs = line.split(' ')
    # print(linesubs)
    index=0
    for idx1,str in enumerate(linesubs):
      # print("Index",linesubs.index(str),"STRING",str, "CHECK")
      if(str!=""):
        index = idx1
        break
    timestamp = linesubs[index]
    # print(timestamp)
    event = linesubs[index+2]
    # print(event)
    index = linesubs[index+3].split('\n')[0]
    print(index)
    entry = [timestamp, event,index]
    print(entry)
    arr.append(entry)
```

```python
df = pd.DataFrame(arr, columns = ['timestamp','event','index'])

df
```
